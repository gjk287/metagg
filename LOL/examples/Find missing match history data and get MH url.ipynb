```python
import os
import sys
import selenium
import glob
import pandas as pd
sys.path.append(os.path.join(os.path.dirname(os.path.abspath('')), 'src'))
# from database import DB
from utils import GamepediaDict
```

```python
save_attribute_error = pd.DataFrame(columns=['year', 'season', 'league_name', 'week', 'team_1', 'team_2', 'set_number'])
count_idx = 0
for file in glob.glob(r'C:\Users\jjames\Dropbox\Cloud_Data\Projects\Game-Data-Platform\meta.gg\LOL\datasets\RawData\MatchHistory\player\*csv'):
    temp_df = pd.read_csv(file)
    temp_df = temp_df[temp_df['state']=='AttributeError'].reset_index(drop=True)
    for idx, val in temp_df.iterrows():
        save_attribute_error.loc[count_idx, ['year', 'season', 'league_name', 'week', 'team_1', 'team_2', 'set_number']] = tuple(val[['year', 'season', 'league_name', 'week', 'team_1', 'team_2', 'set_number']])
        count_idx += 1
    
```

```python
save_attribute_error
```

```python
gp_dict = GamepediaDict()
```

```python
save_missing_mh = pd.DataFrame()

for file in glob.glob(r'C:\Users\jjames\Dropbox\Cloud_Data\Projects\Game-Data-Platform\meta.gg\LOL\datasets\RawData\Gamepedia\match_history_url\*csv'):
    if ('missing' in file) or ('Missing' in file):
        continue
    temp_df = pd.read_csv(file)
    league = file.split('\\')[-1].split('-')[0]
    year = file.split('\\')[-1].split('-')[1]
    season = file.split('\\')[-1].split('-')[2].split('.csv')[0]
    gp_dict.from_year_season_league(year, season, league)
    temp_df[['team_1', 'team_2']] = temp_df[['team_1', 'team_2']].replace(gp_dict.get_dict('team'))
    temp_df[['team_1', 'team_2']] = temp_df[['team_1', 'team_2']].fillna(method='ffill')
    
    atr_err_df = save_attribute_error[(save_attribute_error['league_name']==league) & (save_attribute_error['year']==int(year)) & (save_attribute_error['season']==season)]
    merged_df = pd.merge(atr_err_df, temp_df, how='left', left_on=['year', 'season', 'league_name', 'week', 'team_1', 'team_2', 'set_number'], right_on=['year', 'season', 'league_name', 'week', 'team_1', 'team_2', 'set_number'])
    save_missing_mh = pd.concat([save_missing_mh, merged_df])

    
```

```python
save_missing_mh = save_missing_mh[save_missing_mh['league_name']!='LDL']
```

```python
save_missing_mh.to_csv(r'C:\Users\jjames\Dropbox\Cloud_Data\Projects\Game-Data-Platform\meta.gg\LOL\datasets\RawData\Gamepedia\match_history_url\Missing-Data.csv', index=False)
```
